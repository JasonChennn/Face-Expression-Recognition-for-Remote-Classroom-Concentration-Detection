```python
import numpy as np
import argparse
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
os.environ['CUDA_VISIBLE_DEVICES']='2, 3'
#Info
postive_count = 0
negative_count = 0
```

```python
# Create the model
model = Sequential()

model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(48,48,1)))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(8, activation='softmax'))
print("開始進行推論,結束時請按'q'來獲取專心程度結果.")
model.load_weights('/tmp/model.h5')

# prevents openCL usage and unnecessary logging messages
cv2.ocl.setUseOpenCL(False)

# dictionary which assigns each label an emotion (alphabetical order)
#emotion_dict = {0: "Angry", 1: "Disgusted", 2: "Fearful", 3: "Happy", 4: "Neutral", 5: "Sad", 6: "Surprised"}
emotion_dict = {0: "Absent(-3)", 1: "Concentrate(1)", 2: "Concentrate(3)", 3: "Absent(-5)", 4: "Concentrate(5) ", 5: "Absent(-1)", 6: "Concentrate(1)", 7: "Absent(-1)"}

```

```python
def point(x):
    return {
        3: 5,
        4: 5,
        0: 3,
        2: 3,
        1: 1,
        5: 1,
        6: 1,
        7: 1,
    }[x]
# start the webcam feed
cap = cv2.VideoCapture("/tmp/demo.mp4")
while True:
    # Find haar cascade to draw bounding box around face
    ret, frame = cap.read()
    if not ret:
        break
    facecasc = cv2.CascadeClassifier('/tmp/haarcascade_frontalface_default.xml')
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = facecasc.detectMultiScale(gray,scaleFactor=1.3, minNeighbors=5)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y-50), (x+w, y+h+10), (255, 0, 0), 2)
        roi_gray = gray[y:y + h, x:x + w]
        cropped_img = np.expand_dims(np.expand_dims(cv2.resize(roi_gray, (48, 48)), -1), 0)
        prediction = model.predict(cropped_img)
        maxindex = int(np.argmax(prediction))
        cv2.putText(frame, emotion_dict[maxindex], (x+20, y-60), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)

        if(maxindex == 1 or maxindex == 2 or maxindex == 4 or maxindex == 6):
            postive_count = postive_count + point(maxindex)
        else:
            negative_count = negative_count + point(maxindex)
        print("人臉標籤Index:" + str(maxindex) + "專心分數:" + str(postive_count) + "不專心分數:" + str(negative_count))

    cv2_imshow(cv2.resize(frame,(100,90),interpolation = cv2.INTER_CUBIC))
    if cv2.waitKey(1) & 0xFF == ord('q'):
      break

cap.release()
labels = 'Concentrate', 'Absent'
sizes = [postive_count/(postive_count+negative_count),negative_count/(postive_count+negative_count)]
explode = (0, 0.1) 
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
shadow=True, startangle=90)
if postive_count >= negative_count:
    plt.title('Congrats! You are concentrate.')
else:
    plt.title('Badly! You are absent.')
plt.axis('equal') 
plt.show()
```
